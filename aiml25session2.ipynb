```python
import numpy as np
img = np.zeros((200,200,3), dtype="uint8") #size of image
```

```python
img[:,:,0] = 0
img[:,:,1] = 220
img[:,:,2] = 0
```

```python
import imageio as io
io.imwrite('img.png', img)
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
plt.show()
```
